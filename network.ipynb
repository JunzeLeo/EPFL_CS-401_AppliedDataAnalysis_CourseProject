### Convert involved alias to the PersonId

```python
alias = pd.read_csv(data_folder + 'Aliases.csv')
```

```python
alias_dict = dict(zip(alias.Alias, alias.PersonId))
alias_dict[' h']= alias_dict['h']
alias_dict['duk soo han']= alias_dict['duksoo han']
alias_dict['ki moon ban']= alias_dict['kimoon ban']
alias_dict['rosemarie.howe ']= alias_dict['rosemarie.howe']
alias_dict['cheryl.mills ']= alias_dict['cheryl.mills']
```

```python
def alias2id(alias):
    aliases = str(alias).split(';')
    ids = []
    for x in aliases:
        x = x.lower()\
            .replace(',', '')\
            .replace('-', ' ')\
            .replace('  ', ' ')
        if x in alias_dict.keys():
            ids.append(alias_dict[x])
    return ids
```
